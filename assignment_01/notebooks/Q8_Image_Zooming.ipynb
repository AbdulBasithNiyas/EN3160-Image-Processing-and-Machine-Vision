# Q8 - Image Zooming

```python
import numpy as np, cv2 as cv, matplotlib.pyplot as plt
from pathlib import Path

DATA = Path("../data/a1images/a1q5images")
RESULTS = Path("../results/q8")
RESULTS.mkdir(parents=True, exist_ok=True)
```

### For image 01

```python
img_01_path = DATA / "im01.png"
img_01 = cv.imread(str(img_01_path))
assert img_01 is not None, f"Image not found at {img_01_path}"
print("Loaded:", img_01_path, "shape:", img_01.shape, "dtype:", img_01.dtype)
```

```python
img_01_small_path = DATA / "im01small.png"
img_01_small = cv.imread(str(img_01_small_path))
assert img_01_small is not None, f"Image not found at {img_01_small_path}"
print("Loaded:", img_01_small_path, "shape:", img_01_small.shape, "dtype:", img_01_small.dtype)
```

### Define functions for zooming and computing SSD

```python
def zoom(img, variant, scale=4):
    if variant == 'nn':
        return cv.resize(img, None, fx=scale, fy=scale, interpolation=cv.INTER_NEAREST)
    elif variant == 'bilinear':
        return cv.resize(img, None, fx=scale, fy=scale, interpolation=cv.INTER_LINEAR)
    else:
        raise ValueError("Unknown variant. Use 'nn' or 'bilinear'.")
    

def compute_norm_ssd(image1, image2):
    return np.sum((image1 - image2)**2) / image1.size
```

```python
# Perform zooming
scale_factor = 4
zoom_nn_1 = zoom(img_01_small, 'nn', scale=scale_factor)
zoom_bilinear_1 = zoom(img_01_small, 'bilinear', scale=scale_factor)
```

```python
# Compute SSD between original and zoomed images
ssd_nn_1 = compute_norm_ssd(img_01, zoom_nn_1)
ssd_bilinear_1 = compute_norm_ssd(img_01, zoom_bilinear_1)

print(f"SSD (Nearest Neighbor): {ssd_nn_1}")
print(f"SSD (Bilinear): {ssd_bilinear_1}")
```

```python
# Save zoomed images
cv.imwrite(str(RESULTS/"q8_zoom_nn_1.png"), zoom_nn_1)
cv.imwrite(str(RESULTS/"q8_zoom_bilinear_1.png"), zoom_bilinear_1)

# Plot images
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(cv.cvtColor(img_01, cv.COLOR_BGR2RGB))
plt.title("Original")
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(cv.cvtColor(zoom_nn_1, cv.COLOR_BGR2RGB))
plt.title(f"Nearest Neighbor (ssd = {ssd_nn_1:.2f})")
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(cv.cvtColor(zoom_bilinear_1, cv.COLOR_BGR2RGB))
plt.title(f"Bilinear (ssd = {ssd_bilinear_1:.2f})")
plt.axis('off')
plt.tight_layout()
plt.savefig(RESULTS/"q8_zoom_comparison_1.png", dpi=150)
plt.show()
```
